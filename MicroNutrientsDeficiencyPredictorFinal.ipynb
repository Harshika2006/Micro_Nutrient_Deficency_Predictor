# Importing Basic Lib

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import LabelEncoder
import joblib
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.metrics import accuracy_score
from sklearn.tree import DecisionTreeClassifier
from translate import Translator
```

# Defining Dataset and Making input and output

```python
data=pd.read_csv("D:/AI/DIET PROJECT/micronutrient_deficiency_dataset_full.csv")
uni_sym1=data["Symptom1"].unique()
uni_sym2=data["Symptom2"].unique()
uni_sym3=data["Symptom3"].unique()
uni_def=data["Deficiency"].unique()
le=LabelEncoder()
data["Gender"]=le.fit_transform(data["Gender"])
data["Symptom1"]=le.fit_transform(data["Symptom1"])
data["Symptom2"]=le.fit_transform(data["Symptom2"])
data["Symptom3"]=le.fit_transform(data["Symptom3"])
data["Deficiency"]=le.fit_transform(data["Deficiency"])
x=data.loc[:,["Age","Gender","Symptom1","Symptom2","Symptom3"]]
y=data.loc[:,"Deficiency"]
```

```python
x1=data.loc[:,["Age","Gender","Symptom1","Symptom2"]]
y1=data.loc[:,"Deficiency"]
```

# Spliting of data

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=.80,random_state=20)
x1_train,x1_test,y1_train,y1_test=train_test_split(x1,y1,train_size=.80,random_state=20)
```

# For best accuracy

```python
model=DecisionTreeClassifier()
model_1=DecisionTreeClassifier()
```

```python
para={
    "criterion":["gini","entropy","log_loss"],
    "splitter":["best","random"],
    "max_depth":[1,2,3,4,5,6]
}
dtModel=GridSearchCV(model,param_grid=para,cv=3,scoring="accuracy")
dtModel_1=GridSearchCV(model_1,param_grid=para,cv=3,scoring="accuracy")
dtModel.fit(x_train,y_train)
dtModel_1.fit(x1_train,y1_train)
```

```python
dtModel.best_params_
```

```python
dtModel_1.best_params_
```

```python
y_pred=dtModel.predict(x_test)
y1_pred=dtModel_1.predict(x1_test)
```

```python
print("Testing accuracy:",accuracy_score(y_test,y_pred))
print("Testing accuracy:",accuracy_score(y1_test,y1_pred))
```
